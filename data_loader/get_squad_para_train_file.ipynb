```python
import numpy as np
import pandas as pd
import json
import codecs
from tqdm import tqdm
from datetime import datetime
```

```python
input_path = "/scratch/scratch8/madhurjindal/ACS-QG-Scratch/Datasets/original/SQUAD1.1/dev-v1.1.json"
max_plength = 500
min_plength = 10
```

```python
def normalize_text(text):
    """
    Replace some special characters in text.
    """
    # NOTICE: don't change the text length.
    # Otherwise, the answer position is changed.
    text = text.replace("''", '" ').replace("``", '" ').replace("\t", " ")
    return text
```

```python
start = datetime.now()
num_examples = 0
raw_examples = []
with codecs.open(input_path, "r", encoding="utf8") as infile:
    source = json.load(infile)
    pid = 0
    for article in tqdm(source['data']):
        for para in article['paragraphs']:
            context = para['context']
            context = normalize_text(context)
            for qa in para['qas']:
                question = qa['question']
                question = normalize_text(question)
                for ans_dict in qa['answers']:
                    answer_text = ans_dict['text']
                    answer_text = normalize_text(answer_text)
                    answer_start = ans_dict['answer_start']
                    example = {
                        "question": question,
                        "ans_sent": context,
                        "answer_text": answer_text,
                        "answer_start": answer_start,
                    }
                    raw_examples.append(example)
                    num_examples += 1

print(("Time of get raw examples: {}").format(datetime.now() - start))
print("Number of raw examples: ", len(raw_examples))        
```
